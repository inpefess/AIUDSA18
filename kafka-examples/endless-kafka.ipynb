```python
MAIN_TOPIC = "input_topic"
```

```python
from kafka import KafkaAdminClient
from kafka.admin import NewTopic

admin_client = KafkaAdminClient(**params)
admin_client.create_topics([
    NewTopic(
        name=MAIN_TOPIC,
        num_partitions=1,
        replication_factor=3,
        topic_configs={"retention.bytes": 2 ** 30}
    )
])
admin_client.create_topics([
    NewTopic(
        name="output_topic",
        num_partitions=1,
        replication_factor=3,
        topic_configs={"retention.bytes": 2 ** 30}
    )
])
print(admin_client.list_topics())
```

```python
from kafka import KafkaProducer
import random
from time import sleep

producer = KafkaProducer(**params)
# to stop the stream, interrupt the kernel
while True:
    # this line makes event happen in unpredictable intervals
    sleep(random.randint(1, 10) / 100)
    producer.send(
        MAIN_TOPIC,
        str(random.random()).encode("utf-8")
    )
```
